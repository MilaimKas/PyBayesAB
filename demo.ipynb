# Bernoulli

```python
from conj_priors import Bernoulli
import numpy as np

# Create Likelihhod = Bernoulli object
Bern_test = Bernoulli.Bays_bernoulli()

# create data
p = 0.3
for n in range(50):
    n_trial = np.random.randint(1,10)
    Bern_test.add_rand_experiment(n_trial, p)

# plot total results (cumulative experiments)
Bern_test.plot_tot()

# plot updated posterior after each experiments
Bern_test.plot_exp(type="1D")

# plot updated posterior after each experiments on a plane
Bern_test.plot_exp(type="2D", pi=0, pf = 0.5)

# watch animation
anim = Bern_test.plot_anim()
anim
```

# Poisson

```python
from conj_priors import Poisson
import numpy as np

# Create Likelihhod = Poisson object
Pois_test = Poisson.Bays_poisson()

# create data
mu = 5
n_events = 30
Pois_test.add_rand_experiment(n_events, mu)

# plot total results (cumulative experiments)
Pois_test.plot_tot(muf=10)

# plot updated posterior after each experiments
Pois_test.plot_exp(type="1D", muf=10)

# plot updated posterior after each experiments on a plane
Pois_test.plot_exp(type="2D", muf=10)

# watch animation
anim = Pois_test.plot_anim(muf=15)
anim
```

# Normal

```python
import helper
import numpy as np
from conj_priors import Normal
import matplotlib.pyplot as plt


N = Normal.Bays_norm()
mu = 40
sig = 10

for i in range(20):
    n = np.random.randint(5,50)
    s = sig + np.random.uniform(-1, 1)
    m = mu + np.random.uniform(-1,1)
    N.add_rand_experiment(n, m, s)
```
